# Data Cleaning

```python
import pandas as pd
saliva = pd.read_csv('saliva.tsv', sep='\t')
```

```python
saliva.head(5)
```

```python
saliva_filtered = saliva.copy()
```

```python
def check_covid_status(row):
    if row=='Never Positive':
        return '-'
    else:
        return '+'
```

```python
saliva_filtered['covid_status'] = saliva_filtered['symptom_severity'].apply(check_covid_status)
```

```python
saliva_filtered = saliva_filtered[['sample_name', 'sex','cohort', 'symptom_severity','covid_status']]
```

```python
saliva_filtered
```

```python
saliva_filtered.to_csv('filtered_saliva.csv')
```
